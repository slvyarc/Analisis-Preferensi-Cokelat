**Data Preprocessing**

```python
df
```

```python
# mengubah nama kolom di indeks tertentu
df = df.rename(columns=
 {df.columns[2]: 'Jenis Cokelat Fav',
  df.columns[3]: 'Waktu Konsumsi',
  df.columns[4]: 'Faktor Utama',
  df.columns[5]: 'Rasa Cokelat Fav',
  df.columns[6]: 'Cek Informasi Giji',
  df.columns[7]: 'Mencoba Rasa Baru',
  df.columns[8]: 'Pengaruh Rating dan Ulasan',
  df.columns[9]: 'Mempengaruhi Perasaan',
  })
display(df)
```

```python
# regular expression
import re
# menghapus karakter selain char pada kolom Mempengaruhi Perasaan
df['Mempengaruhi Perasaan'] = df['Mempengaruhi Perasaan'].replace(r'[^a-zA-Z\s.]', '', regex=True)
# menghapus multispasi pada kolom Mempengaruhi Perasaan
df['Mempengaruhi Perasaan'] = df['Mempengaruhi Perasaan'].replace(r'\s+', ' ', regex=True)
```

```python
# mendefinisikan fungsi untuk mengubah teks menjadi Capitalize Each Word
def capitalize_each_word(text):
    words = text.split()
    capitalized_words = [word.capitalize() for word in words]
    new_text = ' '.join(capitalized_words)
    return new_text

# mengubah isi data di kolom 'Nama Lengkap' menjadi Capitalize Each Word
df['Nama Lengkap'] = df['Nama Lengkap'].apply(capitalize_each_word)

# mengubah isi data pada kolom 'Mempengaruhi Perasaan' menjadi huruf kapital
df['Mempengaruhi Perasaan'] = df['Mempengaruhi Perasaan'].str.capitalize()

# mengubah isi data pada kolom 'Pengaruh Rating dan Ulasan' menjadi huruf kapital
df['Pengaruh Rating dan Ulasan'] = df['Pengaruh Rating dan Ulasan'].str.capitalize()

df
```
